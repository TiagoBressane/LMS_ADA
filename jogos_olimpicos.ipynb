```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/summer.csv')
display(df.head())
```

```python
contagem_medalhas = df['Medal'].value_counts()
display(contagem_medalhas)
```

```python
medalhas_ano = df.groupby(['Country', 'Year']).size().unstack(fill_value=0)
display(medalhas_ano)
```

```python
media_historica_medalhas = medalhas_ano.mean(axis=1)
display(media_historica_medalhas)
```

```python
top_medias = media_historica_medalhas.sort_values(ascending=False).head(50)
print(top_medias)
```

```python
media_top_medias = top_medias.mean()
print(f"A média das top 50 médias de medalhas por país é: {media_top_medias:.2f}")
```

```python
mediana_top_medias = top_medias.median()
print(f"A mediana das top 50 médias de medalhas por país é: {mediana_top_medias:.2f}")
```

```python
moda_top_medias = top_medias.mode().tolist()
print(f"A moda das top 50 médias de medalhas por país é: {moda_top_medias}")
```

```python
top_medias = media_historica_medalhas.sort_values(ascending=False).head(50)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_medias.index, y=top_medias.values, palette='viridis')
plt.axhline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media_top_medias:.2f}')
plt.axhline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axhline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Top 50 Países com Maior Média de Medalhas ao Longo dos Anos')
plt.xlabel('País')
plt.ylabel('Média de Medalhas')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Boxplot para a distribuição das top 50 médias de medalhas
plt.figure(figsize=(8, 6))
sns.boxplot(y=top_medias, palette='cividis')
plt.axhline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media_top_medias:.2f}')
plt.axhline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axhline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Boxplot das Top 50 Médias de Medalhas por País')
plt.ylabel('Média de Medalhas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(top_medias, kde=True, bins=10)
plt.axvline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media_top_medias:.2f}')
plt.axvline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axvline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Histograma das Top 50 Médias de Medalhas por País com Média, Mediana e Moda')
plt.xlabel('Média de Medalhas')
plt.ylabel('Frequência')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Violino
plt.figure(figsize=(8, 6))
sns.violinplot(y=top_medias, palette='viridis')
plt.axhline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media_top_medias:.2f}')
plt.axhline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axhline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Violin Plot das Top 50 Médias de Medalhas por País com Média, Mediana e Moda')
plt.ylabel('Média de Medalhas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(top_medias, fill=True)
plt.axvline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media_top_medias:.2f}')
plt.axvline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axvline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Densidade das Top 50 Médias de Medalhas por País com Média, Mediana e Moda')
plt.xlabel('Média de Medalhas')
plt.ylabel('Densidade')
plt.legend()
plt.tight_layout()
plt.show()
```

## Análise Geográfica da Distribuição de Medalhas

### Subtarefa:
Explorar a distribuição de medalhas por continente ou região.

**Motivo**:
Para analisar a distribuição geográfica das medalhas, é necessário associar cada país a um continente ou região. Faremos isso manualmente para os países mais frequentes ou utilizando um mapeamento prévio se disponível. Em seguida, agruparemos os dados por continente/região e contaremos as medalhas.

```python
# Para simplificar, vamos focar nos países que já agrupamos e adicionar uma coluna de continente manualmente
# Esta é uma abordagem simplificada e pode não cobrir todos os países
continent_mapping = {
    'USA': 'América do Norte',
    'URS': 'Europa', # Considerado Europa para simplificar (maior parte)
    'GBR': 'Europa',
    'FRA': 'Europa',
    'GER': 'Europa',
    'ITA': 'Europa',
    'AUS': 'Oceania',
    'HUN': 'Europa',
    'SWE': 'Europa',
    'NED': 'Europa',
    'GDR': 'Europa', # Alemanha Oriental
    'CHN': 'Ásia',
    'JPN': 'Ásia',
    'RUS': 'Europa', # Rússia (parte europeia)
    'CAN': 'América do Norte',
    'ROU': 'Europa',
    'NOR': 'Europa',
    'KOR': 'Ásia',
    'POL': 'Europa',
    'DEN': 'Europa',
    # Para os outros países no grupo 'Other', podemos deixá-los como 'Outro Continente' ou tentar mapear mais.
    # Para esta análise inicial, vamos focar nos países agrupados que mapeamos manualmente.
    'Other': 'Outro Continente'
}

df['Continent'] = df['Country_Grouped'].map(continent_mapping).fillna('Outro Continente')

# Contar medalhas por continente
medals_by_continent = df['Continent'].value_counts()
display(medals_by_continent)

# Visualizar a distribuição de medalhas por continente
plt.figure(figsize=(10, 6))
sns.barplot(x=medals_by_continent.index, y=medals_by_continent.values, palette='viridis')
plt.title('Distribuição de Medalhas por Continente')
plt.xlabel('Continente')
plt.ylabel('Número de Medalhas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Comparação de Edições Específicas dos Jogos

### Subtarefa:
Comparar a distribuição de medalhas em edições específicas dos Jogos Olímpicos de Verão.

**Motivo**:
Selecionar algumas edições dos jogos (por exemplo, uma histórica e uma mais recente) e comparar a contagem de medalhas por país, esporte, ou gênero para identificar tendências e mudanças ao longo do tempo.

```python
# Vamos selecionar duas edições para comparar, por exemplo, 1900 (início) e 2012 (mais recente)
year_1 = 1900
year_2 = 2012

# Filtrar os dados para as edições selecionadas
df_year1 = df[df['Year'] == year_1]
df_year2 = df[df['Year'] == year_2]

print(f"Análise para a edição de {year_1}:")
medals_year1 = df_year1['Country_Grouped'].value_counts().head(10)
display(medals_year1)

print(f"\nAnálise para a edição de {year_2}:")
medals_year2 = df_year2['Country_Grouped'].value_counts().head(10)
display(medals_year2)

# Visualizar a comparação (exemplo: top 10 países)
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

sns.barplot(x=medals_year1.index, y=medals_year1.values, palette='viridis', ax=axes[0], hue=medals_year1.index, legend=False)
axes[0].set_title(f'Top 10 Países com Mais Medalhas em {year_1}')
axes[0].set_xlabel('País')
axes[0].set_ylabel('Número de Medalhas')
axes[0].tick_params(axis='x', rotation=45)

sns.barplot(x=medals_year2.index, y=medals_year2.values, palette='viridis', ax=axes[1], hue=medals_year2.index, legend=False)
axes[1].set_title(f'Top 10 Países com Mais Medalhas em {year_2}')
axes[1].set_xlabel('País')
axes[1].set_ylabel('Número de Medalhas')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Motivo**:
Agrupar os dados por ano e gênero para contar o número de medalhas ganhas por homens e mulheres em cada edição dos jogos. Visualizar essa contagem ao longo do tempo para observar a evolução.

```python
# Contar medalhas por ano e gênero
medals_by_year_gender = df.groupby(['Year', 'Gender']).size().unstack(fill_value=0)

display(medals_by_year_gender)

# Visualizar a evolução da contagem de medalhas por gênero ao longo dos anos
medals_by_year_gender.plot(kind='line', figsize=(15, 8), marker='o')
plt.title('Evolução da Contagem de Medalhas por Gênero ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Medalhas')
plt.legend(title='Gênero')
plt.grid(True)
plt.tight_layout()
plt.show()
```
