```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/summer.csv')
display(df.head())
```

```python
contagem_medalhas = df['Medal'].value_counts()
display(contagem_medalhas)
```

```python
medalhas_ano = df.groupby(['Country', 'Year']).size().unstack(fill_value=0)
display(medalhas_ano)
```

```python
media_historica_medalhas = medalhas_ano.mean(axis=1)
display(media_historica_medalhas)
```

```python
top_medias = media_historica_medalhas.sort_values(ascending=False).head(50)
print(top_medias)
```

```python
media_top_medias = top_medias.mean()
print(f"A mÃ©dia das top 50 mÃ©dias de medalhas por paÃ­s Ã©: {media_top_medias:.2f}")
```

```python
mediana_top_medias = top_medias.median()
print(f"A mediana das top 50 mÃ©dias de medalhas por paÃ­s Ã©: {mediana_top_medias:.2f}")
```

```python
moda_top_medias = top_medias.mode().tolist()
print(f"A moda das top 50 mÃ©dias de medalhas por paÃ­s Ã©: {moda_top_medias}")
```

```python
top_medias = media_historica_medalhas.sort_values(ascending=False).head(50)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_medias.index, y=top_medias.values, palette='viridis')
plt.axhline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'MÃ©dia: {media_top_medias:.2f}')
plt.axhline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axhline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Top 50 PaÃ­ses com Maior MÃ©dia de Medalhas ao Longo dos Anos')
plt.xlabel('PaÃ­s')
plt.ylabel('MÃ©dia de Medalhas')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Boxplot para a distribuiÃ§Ã£o das top 50 mÃ©dias de medalhas
plt.figure(figsize=(8, 6))
sns.boxplot(y=top_medias, palette='cividis')
plt.axhline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'MÃ©dia: {media_top_medias:.2f}')
plt.axhline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axhline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Boxplot das Top 50 MÃ©dias de Medalhas por PaÃ­s')
plt.ylabel('MÃ©dia de Medalhas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(top_medias, kde=True, bins=10)
plt.axvline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'MÃ©dia: {media_top_medias:.2f}')
plt.axvline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axvline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Histograma das Top 50 MÃ©dias de Medalhas por PaÃ­s com MÃ©dia, Mediana e Moda')
plt.xlabel('MÃ©dia de Medalhas')
plt.ylabel('FrequÃªncia')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Violino
plt.figure(figsize=(8, 6))
sns.violinplot(y=top_medias, palette='viridis')
plt.axhline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'MÃ©dia: {media_top_medias:.2f}')
plt.axhline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axhline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Violin Plot das Top 50 MÃ©dias de Medalhas por PaÃ­s com MÃ©dia, Mediana e Moda')
plt.ylabel('MÃ©dia de Medalhas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(top_medias, fill=True)
plt.axvline(media_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'MÃ©dia: {media_top_medias:.2f}')
plt.axvline(mediana_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_top_medias:.2f}')
for moda_valor in moda_top_medias:
    plt.axvline(moda_valor, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda_valor:.2f}')
plt.title('Densidade das Top 50 MÃ©dias de Medalhas por PaÃ­s com MÃ©dia, Mediana e Moda')
plt.xlabel('MÃ©dia de Medalhas')
plt.ylabel('Densidade')
plt.legend()
plt.tight_layout()
plt.show()
```

## AnÃ¡lise GeogrÃ¡fica da DistribuiÃ§Ã£o de Medalhas

### Subtarefa:
Explorar a distribuiÃ§Ã£o de medalhas por continente ou regiÃ£o.

**Motivo**:
Para analisar a distribuiÃ§Ã£o geogrÃ¡fica das medalhas, Ã© necessÃ¡rio associar cada paÃ­s a um continente ou regiÃ£o. Faremos isso manualmente para os paÃ­ses mais frequentes ou utilizando um mapeamento prÃ©vio se disponÃ­vel. Em seguida, agruparemos os dados por continente/regiÃ£o e contaremos as medalhas.

```python
# Para simplificar, vamos focar nos paÃ­ses que jÃ¡ agrupamos e adicionar uma coluna de continente manualmente
# Esta Ã© uma abordagem simplificada e pode nÃ£o cobrir todos os paÃ­ses
continent_mapping = {
    'USA': 'AmÃ©rica do Norte',
    'URS': 'Europa', # Considerado Europa para simplificar (maior parte)
    'GBR': 'Europa',
    'FRA': 'Europa',
    'GER': 'Europa',
    'ITA': 'Europa',
    'AUS': 'Oceania',
    'HUN': 'Europa',
    'SWE': 'Europa',
    'NED': 'Europa',
    'GDR': 'Europa', # Alemanha Oriental
    'CHN': 'Ãsia',
    'JPN': 'Ãsia',
    'RUS': 'Europa', # RÃºssia (parte europeia)
    'CAN': 'AmÃ©rica do Norte',
    'ROU': 'Europa',
    'NOR': 'Europa',
    'KOR': 'Ãsia',
    'POL': 'Europa',
    'DEN': 'Europa',
    # Para os outros paÃ­ses no grupo 'Other', podemos deixÃ¡-los como 'Outro Continente' ou tentar mapear mais.
    # Para esta anÃ¡lise inicial, vamos focar nos paÃ­ses agrupados que mapeamos manualmente.
    'Other': 'Outro Continente'
}

df['Continent'] = df['Country_Grouped'].map(continent_mapping).fillna('Outro Continente')

# Contar medalhas por continente
medals_by_continent = df['Continent'].value_counts()
display(medals_by_continent)

# Visualizar a distribuiÃ§Ã£o de medalhas por continente
plt.figure(figsize=(10, 6))
sns.barplot(x=medals_by_continent.index, y=medals_by_continent.values, palette='viridis')
plt.title('DistribuiÃ§Ã£o de Medalhas por Continente')
plt.xlabel('Continente')
plt.ylabel('NÃºmero de Medalhas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
