```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/summer.csv')
display(df.head())
```

```python
contagem_medalhas = df['Medal'].value_counts()
display(contagem_medalhas)
```

```python
medalhas_ano = df.groupby(['Country', 'Year']).size().unstack(fill_value=0)
display(medalhas_ano)
```

```python
media_historica_medalhas = medalhas_ano.mean(axis=1)
display(media_historica_medalhas)
```

```python
top_medias = media_historica_medalhas.sort_values(ascending=False).head(50)
print(top_medias)
```

```python
mean_of_top_medias = top_medias.mean()
print(f"A média das top 50 médias de medalhas por país é: {mean_of_top_medias:.2f}")
```

```python
median_of_top_medias = top_medias.median()
print(f"A mediana das top 50 médias de medalhas por país é: {median_of_top_medias:.2f}")
```

```python
mode_of_top_medias = top_medias.mode().tolist()
print(f"A moda das top 50 médias de medalhas por país é: {mode_of_top_medias}")
```

```python
top_medias = media_historica_medalhas.sort_values(ascending=False).head(50)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_medias.index, y=top_medias.values, palette='viridis')
plt.axhline(mean_of_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {mean_of_top_medias:.2f}')
plt.axhline(median_of_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {median_of_top_medias:.2f}')
for mode_val in mode_of_top_medias:
    plt.axhline(mode_val, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {mode_val:.2f}')
plt.title('Top 50 Países com Maior Média de Medalhas ao Longo dos Anos')
plt.xlabel('País')
plt.ylabel('Média de Medalhas')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Boxplot for the distribution of the top 50 average medals
plt.figure(figsize=(8, 6))
sns.boxplot(y=top_medias, palette='viridis')
plt.axhline(mean_of_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {mean_of_top_medias:.2f}')
plt.axhline(median_of_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {median_of_top_medias:.2f}')
for mode_val in mode_of_top_medias:
    plt.axhline(mode_val, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {mode_val:.2f}')
plt.title('Boxplot das Top 50 Médias de Medalhas por País')
plt.ylabel('Média de Medalhas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(top_medias, kde=True, bins=10)
plt.axvline(mean_of_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {mean_of_top_medias:.2f}')
plt.axvline(median_of_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {median_of_top_medias:.2f}')
for mode_val in mode_of_top_medias:
    plt.axvline(mode_val, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {mode_val:.2f}')
plt.title('Histograma das Top 50 Médias de Medalhas por País com Média, Mediana e Moda')
plt.xlabel('Média de Medalhas')
plt.ylabel('Frequência')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(8, 6))
sns.violinplot(y=top_medias, palette='viridis')
plt.axhline(mean_of_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {mean_of_top_medias:.2f}')
plt.axhline(median_of_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {median_of_top_medias:.2f}')
for mode_val in mode_of_top_medias:
    plt.axhline(mode_val, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {mode_val:.2f}')
plt.title('Violin Plot das Top 50 Médias de Medalhas por País com Média, Mediana e Moda')
plt.ylabel('Média de Medalhas')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(top_medias, fill=True)
plt.axvline(mean_of_top_medias, color='r', linestyle='dashed', linewidth=2, label=f'Média: {mean_of_top_medias:.2f}')
plt.axvline(median_of_top_medias, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {median_of_top_medias:.2f}')
for mode_val in mode_of_top_medias:
    plt.axvline(mode_val, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {mode_val:.2f}')
plt.title('Density Plot das Top 50 Médias de Medalhas por País com Média, Mediana e Moda')
plt.xlabel('Média de Medalhas')
plt.ylabel('Densidade')
plt.legend()
plt.tight_layout()
plt.show()
```
